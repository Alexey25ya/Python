Дана функция f(x) = (sin(x)/2 + cos(x)/2)^2 * sin(x)

1.Определить корни

2.Найти интервалы, на которых функция возрастает

3.Найти интервалы, на которых функция убывает

4.Построить график

5.Вычислить вершину

6.Определить промежутки, на котором f > 0

7.Определить промежутки, на котором f < 0

```python
from sympy.abc import x
from sympy import *
y = trigsimp(((sin(x)/2) + (cos(x)/2)) ** 2 * sin(x))
y
```

## 1.Определить корни

```python
# f=plt.plot(y,(x,0x4*pi))
f=(solveset(y, x))
f
```

## 2.Найти интервалы, на которых функция возрастает

```python
# Чтобы найти промежутки возрастания (убывания) функции, нужно найти производную функции, 
# приравнять к нулю (найти нули функции) и определить знаки получившихся промежутков на числовой прямой.
# Если производная положительна, то функция возрастает, а если производная отрицательна, то функция убывает.

f = sorted(solveset(diff(y, x), x, domain=Interval(-pi/2, 2*pi)))

# f - список из координат x, в которых производная функции равна 0 (точки экстремума)
print('В интервале -pi/2 до 2*pi\n')
for i in range(1, len(f)):
    if y.subs(x, f[i]) > y.subs(x, f[i] + 0.0000001):
        print(f'Функция возрастает на отрезке: [{simplify(f[i-1])}; {simplify(f[i])}]\n')
        plot(y, (x, f[i-1], f[i])) 
    else:
        continue
        
```

## 3.Найти интервалы, на которых функция убывает

```python
print('В интервале -pi/2 до 2*pi\n')
for i in range(0, len(f)):
    if y.subs(x, f[i]) < y.subs(x, f[i] + 0.0000001):
        print(f'Функция убывает на отрезке: [{simplify(f[i-1])}; {simplify(f[i])}]\n')
        plot(y, (x, f[i-1], f[i])) 
    else:
        continue
```

## 4.Построить график

```python
plot(y, (x, -2*pi, 2*pi)) 
```

## 5.Вычислить вершину

```python
max_min = solveset(diff(y, x), x, domain=Interval(0, 2*pi))
max_min = list(max_min)
y_arr = []
x_arr = []
for i in range(0, len(max_min)):
    x_arr.append(simplify(max_min[i]))
    y_arr.append(y.subs(x, simplify(max_min[i])))
max_y_index = y_arr.index(max(y_arr))
print('Вершина в интервале (0, 2*pi) имеет координаты:\n')
print(f'x-координата: {x_arr[max_y_index]}\n')
print(f'y-координата: {y_arr[max_y_index]}\n')
```

## 6.Определить промежутки, на котором f > 0 (в интервале (0;2*pi))

```python
solve_univariate_inequality(y > 0, x, relational=False, domain=Interval(0, 2*pi))
```

## 7.Определить промежутки, на котором f < 0(в интервале (0;2*pi))

```python
solve_univariate_inequality(y < 0, x, relational=False, domain=Interval(0, 2*pi))
```
